### **5) 집합 자료형(set)**: 교집합(intersaction, &), 합집합(union(), |), 차집합(difference(), -)

```python
set_data = set([1, 2, 3, 5]) # list -> set 형변환
```

```python
set_data #딕셔너리인데 키값만 있게 할 수 있음
```

```python
set_data1 = set([1, 2, 3, 'A']) 
```

```python
b = set_data | set_data1
```

```python
b
```

```python
b = set_data & set_data1
b
```

```python
b = set_data - set_data1
b
```

```python
list(b)
```
