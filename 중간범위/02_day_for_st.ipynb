## **조건문과 반복문**

### **1) if/if ~ else / if ~ elif ~ else**



```python
a = 10
if a > 0:
  print('a는 ', a)
else:
  print('a는 0')
```

### **2) 반복문: for, while**

*   for (반복) 변수 in** 리스트/튜플/문자열**:</br>
> 실행할 명령문
*   while 조건식:
> 실행할 명령문</br>

while 문은 조건의 결과가 참인 동안 명령문을 반복해서 수행



```python
for i in [1,2,3,4,5]:
  print(i)
for i in range(10):
  print(i)
```

```python
listA = [1,2,3,4,5]
len(listA)

for i in range(len(listA)):
  print(i)

for i in listA:
  print("list : ", i)

for i in reversed(range(0,10)):
  print("reversed : ", i)
```

```python
for i in enumerate(listA):
  print(i)
```

## **함수**
함수는 **사용자 정의 함수**와 **내장함수**로 구성

```python
def add(a, b):
  return a + b

print(add(1,3))
```

### **3) 모듈과 패키지**</br>
**모듈**은 함수나 변수 또는 객체를 정의하는 클래스를 모아놓은 파일</br>
**패키지**는 모듈을 모아놓은 디렉토리

```python
class abacus():
  def __init__(self):
    self.a = 0
    self.b = 0
  def add(self, a, b):
    return a + b
```

```python
a = abacus()
a.add(10, 10)
```
