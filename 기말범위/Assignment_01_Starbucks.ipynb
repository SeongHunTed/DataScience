#### find_element_by_css_selector는 최신 버전에서 없어진(deprecated) 메소드입니다.
이거 대신에 find_element()를 사용하셔야 합니다.

- By.ID :	태그의 id값으로 추출</br>
- By.NAME :	태그의 name값으로 추출</br>
- By.XPATH :	태그의 경로로 추출</br>
- By.LINK_TEXT :	링크 텍스트값으로 추출</br>
- By.PARTIAL_LINK_TEXT :	링크 텍스트의 자식 텍스트 값을 추출</br>
- By.TAG_NAME	: 태그 이름으로 추출</br>
- By.CLASS_NAME	: 태그의 클래스명으로 추출</br>
- By.CSS_SELECTOR	: css선택자로 추출</br>

```python
from selenium.webdriver.common.by import By

sbContent = wd.find_element(By.CSS_SELECTOR, 'header.loca_search  > h3 > a')
#sbContent = wd.find_element_by_css_selector('div > a') #문제1) 괄호를 채우세요!
sbContent.click()
time.sleep(5)
```

### 자신의 지역 전체를 크롤링하기

```python
#문제6) 코드자 작동하도록 for문 안의 코드를 채우세요!
tot_name = []
tot_addr = []
tot_tel = []
for n in range(len(storeInfo)):
    sInfo = list(storeInfo[n])
    sName = sInfo[1].string
    sAddr = list(sInfo[3])[0]
    sTel = list(sInfo[3])[2]
    tot_name.append(sName.strip())
    tot_addr.append(sAddr.strip())
    tot_tel.append(sTel.strip())

```

### 데이터 프레임 만들기

```python
import pandas as pd
SB_df = pd.DataFrame({'매장명':tot_name, '주소': tot_addr, '전화번호': tot_tel})
SB_df
```
