### **기초 통계 시각화: Box Plot**

```python
data_a = wine[wine['type']=='red'].quality
data_b = wine[wine['type']=='white'].quality

# notch : 중앙값 도드라지게
plt.boxplot([data_a, data_b], notch=False, whis=1.5)
plt.xlabel('Wine Type')
plt.ylabel('Quality')

plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
wine
```

```python
x = wine.drop(['type'], axis = 1)
```

```python
x
```

```python
y = wine['quality']
```

```python
trainX, testX, trainY, testY = train_test_split(x, y, train_size=0.8, random_state = 0)
```

```python
lr = LinearRegression()
```

```python
lr.fit(trainX, trainY)
```

```python
predictY = lr.predict(testX)
```

```python
predictY
```

```python
testY
```

```python
mse = mean_squared_error(testY, predictY)
```

```python
mse
```

```python
RMSE = np.sqrt(mse)
```

```python
RMSE
```
