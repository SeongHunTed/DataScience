```python
CB4 = pd.read_excel('/content/drive/MyDrive/DataScience/CB_LL.xlsx')
CB4
```

```python
map = folium.Map(location = [37.559978, 126.975291], zoom_start=14)
```

```python
map.save('/content/drive/MyDrive/DataScience/map.html')
```

```python
map_CB4 =  folium.Map(location = [37.559978, 126.975291], zoom_start=14)
```

```python
# iterrow() 시험
for i, info in CB4.iterrows():
  folium.Marker(location=[info['Latitude'], info['Longitude'] ], popup = info['store'], icon=folium.Icon(color='blue', icon='star')).add_to(map_CB4)
```

```python
map_CB4.save('/content/drive/MyDrive/DataScience/map_CB4.html')
```
