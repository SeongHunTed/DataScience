# 기초통계

*   평균(mean)
*   확율밀도함수 = 정규분포함수
*   편차: 편차(deviation)는 관측값에서 평균 또는 중앙값을 뺀 것
*   평균 편차
*   제곱합(sum of squares:): 변동(variation)
*   분산(variance): 관측값에서 평균을 뺀 값을 제곱 
*   표준편차: 평균 에 대한 오차, 즉 평균이 m이고, 표준편차가 3이라고 할때, 실제 값은 m+-3 값이라는 것

# 상관분석

**상관분석**<br>
* 두 변수가 어떤 선형적 관계에 있는지를 분석하는 방법
* 두 변수는 서로 독립적이거나 상관된 관계일 수 있는데, 두 변수의 관계의 강도를 상관관계 라고함
* 상관 분석에서는 상관관계의 정도를 나타내는 단위로 모상관 계수 ρ를 사용
* 상관 계수는 두 변수가 연관된 정도를 나타낼 뿐 인과 관계를 설명하지 않으므로 정확한 예측치를 계산할 수는 없음


**상관 계수 ρ**<br>
* 변수 간 관계의 정도(0~1)와 방향(+, -)을 하나의 수치로 요약해주는 지수로 -1에서 +1 사이의 값을 가짐
* 상관 계수 중에서 많이 사용하는 것은 피어슨 상관 계수(Pearson Correlation Coefficient)임

### **결측치 처리**

**어떻게 결측치를 치환 할 것인가?**
*   age의 결측값은 중앙값으로 치환
*   embark_town이 Embarked와 동일 삭제
*   embarked는 최빈값으로 치환
*   deck는 삭제



**age의 결측값은 중앙값으로 치환**

```python
titanic.describe()
```

```python
titanic['age'].median()
```

```python
titanic['age']= titanic['age'].fillna(titanic['age'].median())
```

```python
titanic.isna().sum()
```

**embark_town 삭제**

```python
titanic.drop('embark_town', axis=1, inplace=True)
```

```python
titanic
```

**embarked는 최빈값으로 치환**

```python
titanic['embarked'].value_counts()
```

```python
titanic['embarked']=titanic['embarked'].fillna('S')
# 혹은 결측치 행 삭제
# titanic.dropna(inplace=True)
```

```python
titanic.isna().sum()
```

**deck 최빈값으로 결측치 치환하기**

```python
titanic.drop('deck', axis=1, inplace=True)
```
