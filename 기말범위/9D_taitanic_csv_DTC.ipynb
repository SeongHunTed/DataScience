```python
from sklearn.model_selection import train_test_split # 트레이닝 데이터, 테스팅 데이터 분리
from sklearn.tree import DecisionTreeClassifier #의사 결정 트리
from sklearn.metrics import accuracy_score # 정확도 성능 분석
```

```python
x = df.drop(['Survived', 'Name', 'Ticket'], axis = 1)
```

```python
y = df['Survived']
```

```python
x
```

```python
trainX, testX, trainY, testY = train_test_split(x, y, test_size=0.2, random_state=13)
```

```python
trainX
```

```python
trainY
```

```python
DTC = DecisionTreeClassifier()
```

```python
DTC.fit(trainX, trainY) # training
```

```python
predict_Y = DTC.predict(testX)
```

```python
accuracy_score(testY, predict_Y)
```
